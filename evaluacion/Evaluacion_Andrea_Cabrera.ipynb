```python
def cambio_columnas(dataframe):
    """Convierte los nombres de las columnas a minúsculas. Esta función modifica el dataframe utilizado (inplace = True), por eso no tiene return.

    Args:
        dataframe (dataframe): el dataframe del cual queremos cambiar sus columnas
    """
    columnas_nuevas = {col :  col.lower() for col in dataframe.columns}
    dataframe.rename(columns = columnas_nuevas, inplace = True)
```

* Haciendo el análisis exploratorio nos deberíamos haber dado cuenta de que algunas de las variables (danceability, energy, key, loudness, speechiness, acousticness, instrumentalness) no son del tipo que deberían. Esto es debido a que los decimales están establecidos como comas y no con puntos. Cread una función que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto.

```python
# Confirmamos que las colummas que deberían ser tipo float son de tipo object:
df[["danceability", "energy", "key", "loudness", "speechiness", "acousticness", "instrumentalness"]].dtypes
```

```python
def cambiar_a_decimal(value):
    """Esta función cambia las comas por puntos

    Args:
        value (str): el string del cual queremos cambiar a float

    Returns:
        float: Devuelve el valor convertido en un float, reemplazando al tipo anterior.
    """
    try: 
        return float(value.replace(",", "."))
    except:
        return np.nan
```

```python
# aplicamos la función a todas las columnas que queremos cambiar el tipo de dato.

lista_col_cambiar = ["danceability", "energy", "key", "loudness", "speechiness", "acousticness", "instrumentalness"]

for col in lista_col_cambiar:
    df[col] = df[col].apply(cambiar_a_decimal)
```

```python
def quita_dolar(value):
    """Esta función elimina el símbolo '$' de la cadena de texto aportada. Este método únicamente modifica si el 'value' es un str.

    Args:
        value (str): el campo del cuál se desea eliminar el símbolo '$'

    Returns:
        str: devuelve la cadena de texto sin el simbolo '$'
    """
    if isinstance(value,str):
        if "$" in value:
            return value.replace("$", "")
        else:
            return value
    else:
        return value
```

* Deberéis contestar a las siguientes preguntas:

* ¿Cuáles son los diez artistas que más han colaborado según los datos que tenemos?

```python
def obtener_colaboradores(cancion):
    """Devuelve un elemento (colaborador) con los siguientes tratamientos previos.
    1. Eliminamos los elementos "(" y ")" del valor de la columna "track"
    2. Creamos una lista separando por "feat." o "featuring" (Pueden ser ambos casos)
    3. Nos quedamos con los elementos del índice 1 en adelante, por si hubiera más elementos en la lista (Esto ocurre)
    4. Eliminamos los espacios de delante y de atrás de cada elemento de la lista (strip)
    5. Eliminamos duplicados (set) creando un Set, y convertimos de vuelta a List (list) quedándonos con el primer elemento.
    PROBLEMA:
    Hay un problema con un registro en el índice 14171. Track tiene el siguente contenido: ""slide (remix) (feat. pop smoke, a boogie wit da hoodie & chris brown) (feat. pop smoke)""
    

    Args:
        cancion (str): Nombre de la canción y colaboradores

    Returns:
        str: Devuelve únicamente los colaboradores
    """
    col_list = re.split("feat\.|featuring", re.sub("\(|\)", "", cancion))[1:]
    final_list = []
    for x in col_list:
        final_list.append(x.strip())
    return list(set(final_list))[0]
    
df_feat["colaboradores"] = df_feat["track"].apply(obtener_colaboradores)
df_feat["colaboradores"].value_counts()
```

```python
colaboradores = df_feat.groupby("colaboradores")['track'].count().sort_values(ascending = False)
df_colaboradores = pd.DataFrame(colaboradores).reset_index()
top_10_colaboradores = df_colaboradores.head(10)
top_10_colaboradores
```

¿Existen valores atípicos (outliers) en alguna categoría de álbumes?

```python
#Comprobamos si hay valores nulos
views_nulas = df_type_views["views"].isnull().sum()
print(f"Visitas con valor nulo: {views_nulas}")
```

```python
def obtener_percentiles(values):
    """Comprobamos si en la columna que pasamos como parámetros hay nulos. En función de eso, realizamos la función acorde a la situación.

    Args:
        values (float): Columna con los datos a analizar

    Returns:
        List[int]: Devuelve una lista de float con los valores de los percentiles 5, 25, 50, 75 y 95.
    """
    valores_nulos = values.isnull().sum()
    if valores_nulos > 0:
        return np.nanpercentile(values, [5, 25, 50 , 75, 95])
    else:
        return np.percentile(values, [5, 25, 50 , 75, 95])
    
def obtener_outliers(values):
    """Obtenemos los valores atípicos leves, es decir, aquellos valores que se encuentran a 1.5 veces los percentiles 25 y 75.
    
    Consulta: https://es.wikipedia.org/wiki/Valor_at%C3%ADpico#Valor_at%C3%ADpico_leve

    Args:
        values (float): Columna con los datos a analizar

    Returns:
        List[int]: Devuelve una lista con los valores límites a partir de los cuáles buscaremos los outliers
    """
    percentiles = obtener_percentiles(values)
    q1, q3 = percentiles[1], percentiles[3]
    rango_interc = q3 - q1
    distancia = 1.5 * rango_interc,
    min_limit = q1 - distancia 
    max_limit = q3 + distancia
    return [int(min_limit), int(max_limit)]
```

```python
df_album = pd.DataFrame(df_type_views[df_type_views["album_type"] == "album"])
df_single = pd.DataFrame(df_type_views[df_type_views["album_type"] == "single"])
df_compilation = pd.DataFrame(df_type_views[df_type_views["album_type"] == "compilation"])

percentiles_album = obtener_percentiles(df_album["views"])
percentiles_single = obtener_percentiles(df_single["views"])
percentiles_compilation = obtener_percentiles(df_compilation["views"])

print(percentiles_album)
print(".........................")
print(percentiles_single)
print(".........................")
print(percentiles_compilation)
```

```python
outliers_album = obtener_outliers(df_album["views"])
outliers_single = obtener_outliers(df_single["views"])
outliers_compilation = obtener_outliers(df_compilation["views"])

print(outliers_album)
print(".........................")
print(outliers_single)
print(".........................")
print(outliers_compilation)
```

Los valores inferiores de las listas anteriores no tienen sentido en nuestro caso porque no podemos tener visitas en negativos, por lo tanto, solo utilizaremos los valores superiores al 1.5 veces al percentil 75.

```python
limit_album = outliers_album[-1] # Ponemos -1 para acceder a la última posición de la lista de los percentiles de las visitas para album.
df_album_outliers = df[(df["album_type"] == "album") & (df["views"] >= limit_album)].sort_values(by="views", ascending = False) # Buscamos el percentil 95 porque es todavía más outlier que el 75, es decir, esos valores estás más alejados de la mediana.
print(f"Número de outliers: {len(df_album_outliers)}")
df_album_outliers.head(2)
```

```python
limit_single = outliers_single[-1] 
df_single_outliers = df[(df["album_type"] == "single") & (df["views"] >= limit_single)].sort_values(by="views", ascending = False)
print(f"Número de outliers: {len(df_single_outliers)}")
df_single_outliers.head(2)
```

```python
limit_compilation = outliers_compilation[-1] 
df_compilation_outliers = df[(df["album_type"] == "compilation") & (df["views"] >= limit_compilation)].sort_values(by="views", ascending = False)
print(f"Número de outliers: {len(df_compilation_outliers)}")
df_compilation_outliers.head(2)
```
